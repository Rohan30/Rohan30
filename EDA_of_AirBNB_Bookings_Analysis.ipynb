### Variables Description 

1. **id** - It provides a unique id to each listing in the dataset. (Numeric)

2. **name** -It describes the property in a concise manner. (Text)

3. **host_id** -It is used to ascertain the ownership of a certain individual. (Numeric)

4. **host_name** -It is the name of the host. (Text)

5. **neighbourhood_group** -categorizes the city into five groups Bronx,Brooklyn,Manhattan,Staten Islands and Queens. (Text)

6. **neighbourhood** - name of neighbourhood within a neighbourhood group. (Text)

7. **latitude** -It Specifies the Latitude of the listing. (Numeric)

8. **longitude**-It denotes the Longitude of the listing. (Numeric)

9. **room_type**-classifies the listing into Private_Room,Shared_Room and Entire_Apartment. (Text)

10. **price** -provides the price of each listed property. (Numeric Continuous)

11. **minimum_nights** -the number of minimum nights one needs to pay for the listing. (Numeric)

12. **number_of_reviews** -denotes the number of reviews received by a listing. (Numeric)

13. **reviews_per_month**-denotes the number of reviews per month. (Numeric Continuous)

14. **calcuated_host_listings_count**-It corresponds to the number of listings hosted by the unique host_id (Numeric)

15. **availability_365**-The number of days the listing is available in a year. (Numeric)

16. **last_review**- The last day of review given for a listing. (Text)

```python
#Finding the number of hosts in each neighbourhood group 
host_areas=df1.groupby('neighbourhood_group')['host_name'].nunique().sort_values(ascending=False).reset_index(name='number_of_hosts')
host_areas
```

```python
#Finding the number of listings for each host
host_listings=df1.groupby('host_name')['calculated_host_listings_count'].count().sort_values(ascending=False).reset_index(name='total listings')
host_listings
```

```python
#creating earnings column
df1['earnings']=df1['price']*df1['minimum_nights']
```

```python
#Top 10 earners
top_10_earners=df1.groupby(['host_name','host_id'])['earnings'].sum().sort_values(ascending=False)[:10].reset_index(name='total_earnings')
top_10_earners
```

```python
#Top 10 least earners
least_top_10_earners=df1.groupby(['host_name','host_id','neighbourhood_group'])['earnings'].sum().sort_values(ascending=True)[:10].reset_index(name='total_earnings')
least_top_10_earners
```

* Here we have tried to find the top 10 most earning hosts of all.
* This survey will help the company to do the behavioural analysis on customer preference while chosing the host. It will allow the company to adept the practices followed by these top 10 host and will apply the same on other host who are not so profitable.


```python
# Most Expensive hosts with least availability
df1.loc[df1['availability_365']==0].sort_values(by='price',ascending =False).groupby(by='neighbourhood_group').head(1)[['name','host_id',	'host_name',	'neighbourhood_group','room_type',	'price']]
```

```python
# Least Expensive hosts with most availability
df1.loc[df1['availability_365']==365].sort_values(by='price',ascending =True).groupby(by='neighbourhood_group').head(1)[['name','host_id',	'host_name',	'neighbourhood_group','room_type',	'price']]
```

```python
#most reviewed hosts
hosts_reviewed=df1.groupby(['host_name'])['number_of_reviews'].sum().sort_values(ascending=False)[:10].reset_index(name='total_reviews')
hosts_reviewed
```

```python
#least reviewed hosts
least_reviewed=df1.groupby(['host_name'])['number_of_reviews'].sum().sort_values(ascending=True)[:10].reset_index(name='total_reviews')
least_reviewed
```

```python
# finding busiest hosts 
busiest_hosts= df1.groupby(['host_name'])['minimum_nights'].count().reset_index()

busy_host= busiest_hosts.sort_values('minimum_nights', ascending=False).head(10)
busy_host
```

```python
#most expensive neighbourhoods 
most_expensive_neighbourhoods= df1.groupby(['neighbourhood','neighbourhood_group','availability_365'])['price'].sum().reset_index()

most_expensive_neighbourhoods= most_expensive_neighbourhoods.sort_values('price', ascending=False).head(10)
most_expensive_neighbourhoods
```

```python
#least expensive neighbourhoods 
least_expensive_neighbourhoods= df1.groupby(['neighbourhood','neighbourhood_group','availability_365'])['price'].sum().reset_index()

least_expensive_neighbourhoods= least_expensive_neighbourhoods.sort_values('price', ascending=True).head(10)
least_expensive_neighbourhoods
```

```python
import statistics
from collections import Counter

#NLP
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

#word cloud
from wordcloud import WordCloud, ImageColorGenerator
sns.set_theme(style="ticks", color_codes=True)
```

```python
#stop words removal:
"""
1) download the nltk stopwords in english.
2) replace pantuation and special symbols.
3) make it string and convert all into lower case.
then split the words and compair each with the stopwords and keep only those which are not in stopword list.
"""
def stop_word_removal(df):
  stop = stopwords.words('english')
  df.name = df.name.str.replace("[^\w\s]", "").str.lower()
  df.name = df.name.apply(lambda x: ' '.join([item for item in x.split() if item not in stop]))
```

```python
text = stop_word_removal(df1)
text = " ".join(str(v) for v in df1.name)
```

```python
#wordcloud generator.
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=300, background_color="white").generate(text)
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#list of words.
name_list = list(df1.name.values)
words = []
for i in name_list:
  words+=i.split()
```

```python
#let's see top 50 used words .
top_50_words=Counter(words).most_common()
top_50_words=top_50_words[0:50]
```
